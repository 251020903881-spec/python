```python
n = int(input("Nhap n: "))
tong = 0
for i in range(n):
    x = float(input("Nhap so: "))
    tong += x
print("Trung binh cong:", tong / n)
```

```python
n = int(input("Nhap so luong phan tu: "))
dem = 0
for i in range(n):
    x = int(input("Nhap so: "))
    if x % 2 == 0:
        dem += 1
print("So luong so chan:", dem)
```

```python
n = int(input("Nhap so luong phan tu: "))
x = int(input("Nhap so: "))
lon_nhat = x
nho_nhat = x

for i in range(n - 1):
    x = int(input("Nhap so: "))
    if x > lon_nhat:
        lon_nhat = x
    if x < nho_nhat:
        nho_nhat = x

print("So lon nhat:", lon_nhat)
print("So nho nhat:", nho_nhat)
```

```python
def giaithua(n):
    if n == 0:
        return 1
    return n * giaithua(n - 1)

n = int(input("Nhap n: "))
print("Giai thua:", giaithua(n))
```
