```python
nam = int(input("Nhập một năm bất kỳ: "))

if (nam % 4 == 0 and nam % 100 != 0) or (nam % 400 == 0):
    print(f"{nam} là năm nhuận.")
else:
    print(f"{nam} không phải là năm nhuận.")
```

```python
n = int(input("Nhập một số nguyên dương n: "))

if n <= 0:
    print("Vui lòng nhập một số nguyên dương.")
elif n > 0:
    # Cách 1: Sử dụng vòng lặp for
    tong_for = 0
    for i in range(1, n + 1):
        tong_for += i
    print(f"Tổng từ 1 đến {n} (sử dụng vòng lặp for) là: {tong_for}")

    # Cách 2: Sử dụng công thức
    tong_cong_thuc = n * (n + 1) // 2
    print(f"Tổng từ 1 đến {n} (sử dụng công thức) là: {tong_cong_thuc}")
```

```python
so_am = 0
so_duong = 0
so_bang_0 = 0

print("Vui lòng nhập 10 số nguyên:")

count = 0
while count < 10:
    try:
        so_str = input(f"Nhập số thứ {count + 1}: ")
        so = int(so_str)

        if so < 0:
            so_am += 1
        elif so > 0:
            so_duong += 1
        else:
            so_bang_0 += 1
        count += 1 # Chỉ tăng biến đếm nếu đầu vào hợp lệ
    except ValueError:
        print("Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")

print(f"Số lượng số âm: {so_am}")
print(f"Số lượng số dương: {so_duong}")
print(f"Số lượng số bằng 0: {so_bang_0}")
```

```python
while True:
    try:
        so_nhap = int(input("Nhập một số nguyên từ 1 đến 9: "))
        if 1 <= so_nhap <= 9:
            break
        else:
            print("Số bạn nhập không nằm trong phạm vi từ 1 đến 9. Vui lòng nhập lại.")
    except ValueError:
        print("Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")

print(f"Bảng cửu chương của {so_nhap}:")
for i in range(1, 11): # Bảng cửu chương thường in đến 10 lần
    print(f"{so_nhap} x {i} = {so_nhap * i}")
```

```python
import math

def kiem_tra_so_nguyen_to(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True

while True:
    try:
        so_nhap = int(input("Nhập một số nguyên dương để kiểm tra số nguyên tố: "))
        if so_nhap > 0:
            break
        else:
            print("Vui lòng nhập một số nguyên dương.")
    except ValueError:
        print("Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")

if kiem_tra_so_nguyen_to(so_nhap):
    print(f"{so_nhap} là số nguyên tố.")
else:
    print(f"{so_nhap} không phải là số nguyên tố.")
```
