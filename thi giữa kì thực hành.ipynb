```python
#bài 1
n = int(input("Nhập n: "))
tong = 0
for i in range(1, n + 1):
    if i % 2 == 0:
        tong += i
print("Tổng các số chẵn là:", tong)
```

```python
s = input("Nhập chuỗi: ")
words = s.split()
word_count = len(words)
if word_count > 0:
    longest_word = ""
    for word in words:
        if len(word) > len(longest_word):
            longest_word = word
    print(f"số từ: {word_count}")
    print(f"từ dài nhất: {longest_word}")
else:
    print("Chuỗi trống!")
```

```python
#bài 3
input_str = input("Nhập danh sách các số: ")
numbers = list(map(int, input_str.split()))
unique_numbers = list(set(numbers))
unique_numbers.sort()
print(f"danh sách sắp xếp: {unique_numbers}")
if len(unique_numbers) < 2:
    print("Không có số lớn thứ hai.")
else:
    second_largest = unique_numbers[-2]
    print(f"số lớn thứ 2: {second_largest}")
```

```python
#bài 4
input_data = input("Nhập danh sách điểm: ")
raw_scores = list(map(float, input_data.split()))
valid_scores = [score for score in raw_scores if 0 <= score <= 10]
if not valid_scores:
    print("Không có điểm hợp lệ để tính toán.")
else:
    average = sum(valid_scores) / len(valid_scores)
    passed_count = len([score for score in valid_scores if score >= 5])
    print(f"điểm trung bình: {average:.1f}")
    print(f"số người đạt: {passed_count}")
```
